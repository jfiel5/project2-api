## Part 2: Explore one Wikipedia page with the beautifulsoup package

```python
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python

#save and print the text content of a page with all tags removed
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd


# Load and parse the Wikipedia page
url = "https://en.wikipedia.org/wiki/2025_4_Nations_Face-Off"
source = requests.get(url).text
soup = BeautifulSoup(source, 'html.parser')

# Get the title
title = soup.title.text
print("Page Title:", title)

# Get all text on the page
text_content = soup.get_text()

# Function to clean the text
def clean_text(text):
    text = text.lower()  # Convert to lowercase
    text = ' '.join(text.split())  # Remove extra spaces and newlines
    return text

# Clean the text
cleaned_text = clean_text(text_content)

# Save raw and cleaned text to files
with open("wiki_output.txt", "w", encoding="utf-8") as f:
    f.write(text_content)

with open("wiki_output_clean.txt", "w", encoding="utf-8") as f:
    f.write(cleaned_text)

# Print the cleaned text
print(cleaned_text)
```

```python
#using the links in the "link" column, download the first 1000 characters of each script
#use requests and bs4, remember to remove all html tags
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Load your CSV file
scripts = pd.read_csv('pudding_data.csv')

# Define a function to clean text and return only the first 1000 characters
def clean(text):
    text = " ".join(text.split())  # Normalize whitespace
    return text[:1000]  # Limit to the first 1000 characters

# Get the links from the "link" column
link_url = scripts["link"].tolist()

# Lists to store results
link_text = []
title = []
count = 0

for idx, url in enumerate(link_url):
    if not url.startswith(("http://", "https://")):
        url = "https://" + url.lstrip("/")  # Prepend https:// if missing

    try:
        # Fetch the page content
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Raise error for bad responses

        # Parse the HTML using BeautifulSoup
        soup = BeautifulSoup(response.text, "html.parser")
        text = clean(soup.get_text())  # Clean and limit text to 1000 characters

        # Get the title of the page or use a default name
        page_title = clean(soup.title.text) if soup.title else f"script_{idx}"

        # Append to the lists
        link_text.append(text)
        title.append(page_title)


    except Exception as e:
        link_text.append(f"Error: {e}")
        title.append("Error")

```

```python
scripts["text"]=link_text
```

```python
#add a new column to the df with the text downloaded
#save this new dataframe as "pudding_texts.csv"
pudding_texts=pd.DataFrame(scripts)
pudding_texts
```
